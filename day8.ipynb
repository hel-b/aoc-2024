```python
from operator import itemgetter
test = [
    "............",
    "........0...",
    ".....0......",
    ".......0....",
    "....0.......",
    "......A.....",
    "............",
    "............",
    "........A...",
    ".........A..",
    "............",
    "............",
]


def solve(data):
    # get all unique character types
    tmp_str = ""
    for line in data:
        tmp_str += line
    types = set(tmp_str.replace(".", ""))

    # loop through all unique character types
    ants = {
        type: [
            (i, j)
            for i, line in enumerate(data)
            for j, c in enumerate(line)
            if c == type
        ]
        for type in types
    }

    row = len(data)
    col = len(data[0])

    def in_bounds(pt):
        return 0 <= pt[0] < row and 0 <= pt[1] < col

    antis = []
    for ant in ants:
        for i, pnt1 in enumerate(ants[ant][: len(ants[ant]) - 1]):
            for j, pnt2 in enumerate(ants[ant][i + 1 :]):
                delta = (pnt1[0] - pnt2[0], pnt1[1] - pnt2[1])
                pos_antis = [
                    (pnt1[0] + delta[0], pnt1[1] + delta[1]),
                    (pnt2[0] - delta[0], pnt2[1] - delta[1]),
                ]
                for anti in pos_antis:
                    if in_bounds(anti) and data[anti[0]][anti[1]] != ant:
                        antis.append(anti)

    return len(set(antis))


print(f"part 1 - test: {solve(test)}")
full_data = []
with open("day8.txt") as f:
    full_data = f.read().splitlines()
print(f"part 1 - result: {solve(full_data)}")
```
